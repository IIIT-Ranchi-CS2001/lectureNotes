## Key Features of Pandas:

**1. Data Structures:**

* **Series:** A one-dimensional (1D) ***labeled array***, similar to a column in an Excel sheet or a NumPy array but with additional functionality. It can hold data of any type, such as: integers, strings, python objects etc.



* **DataFrame:** A two-dimensional (2d) table-like structure, similar to a spreadsheet or a SQL table, with labeled rows and columns.



- Each **column** in a `DataFrame` is a `Series` 


**2. Built on NumPy:**

* Pandas is built on top of the NumPy library, so it inherits the efficient performance of NumPy arrays.
* It handles missing data gracefully, unlike NumPy arrays, which require workarounds.

**3. Integration with Python Ecosystem:**

* Works seamlessly with other popular Python libraries like matplotlib (visualization) and scikit-learn (machine learning).

## Why Use Pandas?
Pandas simplifies many aspects of data handling, making it invaluable for anyone working with data. Some highlights include:

**1. Ease of Use:**

* Provides an intuitive API to load, transform, and analyze data with minimal code.

**2. Powerful Data Manipulation:**

* Allows for tasks like filtering, grouping, merging, reshaping, and pivoting data with ease.

## Aggregating statistics grouped by category

**Grouping:**  

Grouping involves splitting the data into groups based on a column or set of columns, and then applying some computation on each group.

basically:
1. Split the data into groups

2. Apply a function to each group independently

3. Combine the results into a data structure




##### What is the average age for male versus female Titanic passengers?

```python
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

Explanation:  
- We wanted the average age for each gender, a subselection on these two columns is made first: `titanic[["Sex", "Age"]]`.
- Next, the `groupby()` method is applied on the Sex column to make a group per category. The average age for each gender is calculated and returned.

```python
titanic.groupby("Sex")["Age"].mean()
```

#### Grouping can be done by multiple columns at the same time. Provide the column names as a list to the `groupby()` method.

##### What is the mean ticket fare price for each of the sex and cabin class combinations?

```python
titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
```

## **Merging and Concatenation**

In Pandas, merging and concatenation are essential tools for combining datasets. While they serve similar purposes, they work differently and are suited to different scenarios.

### 1. Merging in Pandas
The `.merge()` function in Pandas is similar to SQL joins. It combines two `DataFrames` based on a **common key (column)**.

**Key Types of Merge:**
1. **Inner Join:** Returns only the rows with matching keys in both DataFrames.
2. **Left Join:** Returns all rows from the left DataFrame and matching rows from the right.
3. **Right Join:** Returns all rows from the right DataFrame and matching rows from the left.
4. **Outer Join:** Returns all rows from both DataFrames, filling in missing values where there are no matches.


```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Doremon', 'Sabu', 'Chacha']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 4],
    'Salary': [7000, 8000, 9000]
})

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Inner Join (default)
merged_inner = pd.merge(df1, df2, on='ID')
print("\nInner Join:")
print(merged_inner)

# Left Join
merged_left = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Join:")
print(merged_left)

# Outer Join
merged_outer = pd.merge(df1, df2, on='ID', how='outer')
print("\nOuter Join:")
print(merged_outer)
```

### 2. Concatenation in Pandas
The `.concat()` function in Pandas combines two or more `DataFrames` vertically (row-wise) or horizontally (column-wise).  
* It does not require a common key.
* Can be horizontal (axis=1) or vertical (axis=0)

```python
df1 = pd.DataFrame({
    'Name': ['Nagaraj', 'Mogambo'],
    'Age': [95, 80]
})

df2 = pd.DataFrame({
    'Name': ['Spider', 'Doremon'],
    'Age': [45, 20]
})

print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Vertical Concatenation
concat_vertical = pd.concat([df1, df2], ignore_index = True)
print("\nVertical Concatenation:")
print(concat_vertical)

# Horizontal Concatenation
concat_horizontal = pd.concat([df1.reset_index(drop=True), df2.reset_index(drop=True)], axis=1)
print("\nHorizontal Concatenation:")
print(concat_horizontal)
```

## Some Advanced Topics**
### **1. Time Series**

Pandas excels at working with time-series data. Time-series analysis typically involves handling dates, timestamps, and performing date-based operations.

```python
# Creating a time series DataFrame
date_range = pd.date_range(start='2023-01-01', end='2023-01-07', freq='D')
data = {'Date': date_range, 'Temperature': [30, 32, 31, 29, 35, 28, 27]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Setting 'Date' as the index
df.set_index('Date', inplace=True)

# Filtering data for a specific range
filtered = df.loc['2023-01-02':'2023-01-05']
print("\nFiltered Data:")
print(filtered)

# Resampling data to weekly mean
resampled = df.resample('W').mean()
print("\nWeekly Mean Temperature:")
print(resampled)
```

### 2. Pivot Tables in Pandas
Pivot tables summarize data by aggregating it based on rows and columns. They are excellent for cross-tabulations.

```python
data = {
    'City': ['New York', 'New York', 'Los Angeles', 'Los Angeles', 'Chicago'],
    'Date': ['2023-01-01', '2023-01-02', '2023-01-01', '2023-01-02', '2023-01-01'],
    'Sales': [200, 250, 300, 400, 150],
    'Category': ['Electronics', 'Clothing', 'Clothing', 'Electronics', 'Electronics']
}
df = pd.DataFrame(data)

# Creating a pivot table
pivot = df.pivot_table(
    values='Sales', 
    index='City', 
    columns='Category', 
    aggfunc='sum', 
    fill_value=0
)
print("Pivot Table:")
print(pivot)
```

### Bar Plot with Pivot Table

```python
#import matplotlib.pyplot as plt

# Bar plot for pivot table
pivot.plot(kind='bar', title='City-wise Sales by Category', figsize=(8, 4))
plt.xlabel('City')
plt.ylabel('Sales')
plt.show()
```
