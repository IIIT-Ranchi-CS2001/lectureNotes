## 2. Matplotlib Object Hierarchy


There is an Object Hierarchy within Matplotlib. In Matplotlib, a plot is a hierarchy of nested Python objects.  

A **hierarchy** means that there is a tree-like structure of Matplotlib objects underlying each plot.  

A **Figure** object is the outermost container for a Matplotlib plot. The **Figure** object contain multiple **Axes** objects. So, the **Figure** is the final graphic that may contain one or more **Axes**. The **Axes** represent an individual plot.  


So, we can think of the **Figure** object as a box-like container containing one or more **Axes**. The **Axes** object contain smaller objects such as tick marks, lines, legends, title and text-boxes.

## 9.	Object-Oriented API


The **Object-Oriented API** is available for more complex plotting situations. It allows us to exercise more control over the figure. In Pyplot API, we depend on some notion of an "active" figure or axes. But, in the **Object-Oriented API** the plotting functions are methods of explicit Figure and Axes objects.


**Figure** is the top level container for all the plot elements. We can think of the **Figure** object as a box-like container containing one or more **Axes**. 


The **Axes** represent an individual plot. The **Axes** object contain smaller objects such as axis, tick marks, lines, legends, title and text-boxes.


The following code produces sine and cosine curves using Object-Oriented API.

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1), 'b-')
ax[1].plot(x1, np.cos(x1), 'r-');
```

### Objects and Reference


The main idea with the **Object Oriented API** is to have objects that one can apply functions and actions on.   
The real advantage of this approach becomes apparent when more than one figure is created or when a figure contains more than one subplot.


We create a reference to the figure instance in the **fig** variable.   
Then, we ceate a new axis instance **axes** using the 
**add_axes** method in the Figure class instance fig as follows:-



```python
fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

#plt.plot(x2, y2)

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x2, y2, '.')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')
```

## Error Bar Chart



In experimental design, the measurements lack perfect precision. So, we have to repeat the measurements. It results in 
obtaining a set of values. The representation of the distribution of data values is done by plotting a single data point 
(known as mean value of dataset) and an error bar to represent the overall distribution of data.


We can use Matplotlib's **errorbar()** function to represent the distribution of data values. It can be done as follows:-

```python
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')
#plt.axis([-0.5, 4, 0, 1.2])
plt.grid(True)
plt.show();
```

## Stacked Bar Chart


We can draw stacked bar chart by using a special parameter called **bottom** from the plt.bar() function.

```python
A = [15., 30., 45., 22.]
B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)

plt.show()
```

The optional **bottom** parameter of the plt.bar() function allows us to specify a starting position for a bar. Instead of running from zero to a value, it will go from the bottom to value. The first call to plt.bar() plots the blue bars. The second call to plt.bar() plots the red bars, with the bottom of the red bars being at the top of the blue bars.

**Key Components of a Boxplot:**
* **Box:** Represents the interquartile range (IQR) - the middle 50% of the data.
* **Whiskers:** Extend to the minimum and maximum values that are not considered outliers.
* **Median Line:** The line inside the box represents the median of the data.
* **Outliers:** Data points outside the whiskers, plotted as individual points.

## Contour Plot


**Contour plots** are useful to display three-dimensional data in two dimensions using contours or color-coded regions.
**Contour lines** are also known as **level lines** or **isolines**. **Contour lines** for a function of two variables are 
curves where the function has constant values. They have specific names beginning with iso- according to the nature of the variables being mapped.


There are lot of applications of **Contour lines** in several fields such as meteorology(for temperature, pressure, rain, wind
speed), geography, magnetism, engineering, social sciences and so on.


The density of the lines indicates the **slope** of the function. The **gradient** of the function is always perpendicular to the contour lines. When the lines are close together, the length of the gradient is large and the variation is steep.


A **Contour plot** can be created with the **plt.contour()** function as follows:-

```python
# Create a matrix
matrix1 = np.random.rand(10, 20)

cp = plt.contour(matrix1)

plt.show()
```

## Example of 3D plot

## Few more examples of Object-Oriented API


The Object-Oriented (OO) API provides more control and flexibility over the plots, making it more suitable for complex and customized visualizations.

### 1. Creating Figures and Axes
In the OO approach, you create a `Figure` object and one or more `Axes` objects. Each `Axes` object is a plot or graph inside the figure.

```python
import matplotlib.pyplot as plt

# Create a figure and an Axes
fig, ax = plt.subplots()

# Plot some data on the Axes
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Set title and labels using ax
ax.set_title('Sample Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Show the plot
plt.show()
```

### 2. Multiple Axes (Subplots)
To create multiple subplots in the same `figure`, you can use `plt.subplots()` to create multiple `Axes`x objects in a grid.

```python
# Create a 2x2 grid of Axes
fig, axs = plt.subplots(2, 2)

# First subplot
axs[0, 0].plot([1, 2, 3], [1, 4, 9])
axs[0, 0].set_title('Square Numbers')

# Second subplot
axs[0, 1].plot([1, 2, 3], [1, 8, 27])
axs[0, 1].set_title('Cube Numbers')

# Third subplot
axs[1, 0].plot([1, 2, 3], [1, 16, 81])
axs[1, 0].set_title('Fourth Power')

# Fourth subplot
axs[1, 1].plot([1, 2, 3], [1, 32, 243])
axs[1, 1].set_title('Fifth Power')

# Automatically adjust layout
plt.tight_layout()
plt.show()
```

### 3. Using Methods on Axes
In the OO interface, you directly call methods on `ax` objects to create plots, set labels, or modify the axes. For example, to change the limits of the axes or add a grid:

```python
fig, ax = plt.subplots()

# Plot data
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Set the x and y axis limits
ax.set_xlim([0, 5])
ax.set_ylim([0, 35])

# Add a grid
ax.grid(True)

# Display the plot
plt.show()
```

### 4. Adding Multiple Plots to the Same Axes
You can add multiple lines or scatter plots to the same axes by calling `plot()` or `scatter()` multiple times on the same `ax` object.

```python
fig, ax = plt.subplots()

# First plot
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], label='Line 1')

# Second plot on the same axes
ax.plot([1, 2, 3, 4], [30, 25, 20, 10], label='Line 2')

# Third scatter plot on the same axes
ax.scatter([1, 2, 2.5, 3, 4], [5, 15, 22.5, 24, 27], label='Line 3', marker = '*', color = 'red', s= 80)  #s is size of marker

# Add a legend
ax.legend()

plt.show()
```

### 5. Customizing Ticks and Labels
With the OO approach, you can customize the ticks and labels directly via the `ax` object:


```python
fig, ax = plt.subplots()

# Plot data
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Customizing ticks
ax.set_xticks([1, 2, 3, 4])
ax.set_yticks([10, 15, 20, 25, 30])

# Customizing tick labels
ax.set_xticklabels(['One', 'Two', 'Three', 'Four'])
ax.set_yticklabels(['Ten', 'Fifteen', 'Twenty', 'Twenty Five', 'Thirty'])

plt.show()
```

```python
# Annotations in the OO interface can be added using the annotate() method:
fig, ax = plt.subplots()

# Plot data
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])

# Annotate a specific point
ax.annotate('Highest Point', xy=(4, 30), xytext=(3, 28), 
             arrowprops=dict(facecolor='grey', shrink=0.05))

plt.show()
```
