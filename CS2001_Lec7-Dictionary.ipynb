# Dictionaries

#### A **dictionary** in Python is a collection of key-value pairs. 
#### It is an **ordered (PYTHON 3.7 ONWARDS);** (*thanks Ujjwal for pointing this out*), 
#### **mutable, and indexed** data structure where each ***key is mapped*** to a specific **value**.

##### Key Features of a Dictionary:
* **~Un~Ordered:** Items in a dictionary do not have an index, and their order is not guaranteed (***in Python 3.6 or earlier versions***).
* Since, 3.7, dictionary is **ORDERED**, thus it  preserves the insertion order of the keys. 
* **Mutable:** You can change values, add or remove key-value pairs after a dictionary is created.
* **Indexed by Keys:** Each element in a dictionary is accessed by a unique key.
* **Keys must be unique and immutable:** Keys can be strings, numbers, or tuples, but they must be unique and cannot be lists, sets, or other mutable objects.

### Creating a Dictionary
#### You can create a dictionary using curly braces **{ }** or the **dict()** constructor.

```python
# Using curly braces
student = {
    "name": "Utkarsh",
    "age": 20,
    "department": "CS"
}
print(student)

# Using the dict() constructor
student2 = dict(age=20, name="Utkarsh", department="CS")
# Note, no quotes for the keys because keys are assumed to be strings
print(student2)
```

### Accessing Elements in a Dictionary
### **You can access a value by using its corresponding key.**

```python
# Accessing values
student = {
    "name": "Utkarsh",
    "age": 20,
    "department": "CS"
}
print(student)
a = student["name"]
print(a)
print(student["department"])
```

```python
## we can't access using index like we did for lists & strings, because?? Dictionary is _ _ _ _ _ _?? 
print(student[0])
```

#### Alternatively, use the **get()** method, which returns None (or a specified default) if the key doesn't exist:

```python
print(student.get("age"))      
print(student.get("graduation"))  
print(student.get("graduation", "hello"))  
```

### Adding and Modifying Dictionary Elements
* You can add new key-value pairs or modify existing ones.

```python
# Adding a new key-value pair
student["graduation_year"] = 2027
print(student)

# Modifying an existing value
student["age"] = 23
print(student)
```

### Removing Dictionary Elements
#### You can remove elements using methods like **pop(), popitem(), and del**.

```python
# Remove a specific key-value pair
student = {
    "age": 20,
    "name": "Utkarsh",
    "department": "CS"
}
a = student.pop("age")  # Removes the key 'age'
print(student)
print(a)

# Remove the last inserted key-value pair (in Python 3.7+) (LIFO)
b =student.popitem()   # Removes 'graduation_year'
print(b)
  
# Remove a key using del
del student["name"]
print(student)
```

### Unpacking a dictionary

### Merging dictionaries using **

```python
# Two dictionaries whose keys are different can be merged using ** operator
FruitsPrice={'Mango':50, 'Apple':100, 'Guava':30}
VeggiePrice={'Onion':25, 'Potato':30, 'Tomato':60}
Pricelist={**FruitsPrice,**VeggiePrice}
print(Pricelist)

#If we try to merge two dictionaries with same keys, only the latter will be inserted.
#( Whenever there is duplication of keys, values of the last group of keys are taken)
```

### Dictionary Methods
* Some commonly used methods in Python dictionaries are:

1. **items( ):** Returns a view object with a list of key-value pairs.
2. **keys( ):** Returns a view object containing the dictionary's keys.
3. **values( ):** Returns a view object with all the values.
4. **clear( ):** Removes all elements from the dictionary.
5. **update( ):** Adds key-value pairs from another dictionary or iterable.

```python
student = {
    "name": "Utkarsh",
    "age": 20,
    "department": "CS"
}

# Dictionary methods examples
print(student.items())    
print(dict(student.items()))
print('---------------------')
print(student.keys())     
keys = list(student.keys())
print(keys)
print('---------------------')
print(tuple(student.values()))   
print('---------------------')
```

```python
# Update dictionary with new key-value pairs
student.update({"GPA": 93.9, "status": "active"})
print(student)
```

```python
student.clear()
print(student)
```

### Dictionary Comprehension
#### Just like list comprehensions, you can create dictionaries using dictionary comprehension.

##### {key_expr: value_expr for var in iterable}

* **key_expr-** This is the expression to geenrate the key(s).
* **:** - you know this is needed for dictionary key:value pair
* **value_expr-** This is the expression to geenrate the value

```python
# Dictionary comprehension example
squares = {x: x*x for x in range(1, 6)}      #
print(squares)  
```

### Iterating Over a Dictionary
* You can loop through dictionaries to access keys, values, or both.

```python
student = {
    "name": "Utkarsh",
    "age": 20,
    "department": "CS"
}
print("===Loop through keys===")
for key in student:
    print(key, student[key])

print("===Loop through key-value pairs===")
for key, value in student.items():
    print(key, value)
```
