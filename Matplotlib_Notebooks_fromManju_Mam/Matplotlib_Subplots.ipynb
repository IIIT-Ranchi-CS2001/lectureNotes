```python
import matplotlib.pyplot as plt
```

```python
import random
fobj=plt.figure(figsize=(6,6),facecolor='#00FF00')
spobj1=fobj.add_subplot(221)
spobj2=fobj.add_subplot(224)
x1=spobj1.set_xticks([n for n in range(0,11)])
y1=spobj1.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,10) for n in range(10)]
Y1=sorted(Y)
Y2=[n/2 for n in Y1]
spobj1.set_xlabel("Time in Seconds")
spobj1.set_ylabel("Amplitude in Volts")
spobj1.set_title("Sample Graphs")
spobj1.plot(X,Y1,color='r',marker='o',linestyle='-',mec='g',mfc='b',label='Full')
spobj1.plot(X,Y2,color='g',marker='D',linestyle=':',mec='b',mfc='r',label='Half')
spobj1.legend()
```

```python
import random
fobj=plt.figure(figsize=(20,10),facecolor='#00FF00')
fobj.suptitle("Electrical Characteristics")
spobj1=fobj.add_subplot(121)
spobj2=fobj.add_subplot(122)
x1=spobj1.set_xticks([n for n in range(0,11)])
y1=spobj1.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,10) for n in range(10)]
Y1=sorted(Y)
Y2=[n/2 for n in Y1]
spobj1.set_xlabel("Time in Seconds")
spobj1.set_ylabel("Voltage in volts")
spobj1.set_title("Voltage Profile")
spobj1.plot(X,Y1,color='r',marker='o',linestyle='-',mec='g',mfc='b',label='Full Load')
spobj1.plot(X,Y2,color='g',marker='D',linestyle=':',mec='r',mfc='k',label='Half load')
spobj1.legend()
x2=spobj2.set_xticks([n for n in range(0,11)])
y2=spobj2.set_yticks([n/2 for n in range(0,11)])
X=[n for n in range(10)] 
Y=[random.random()*n/2 for n in range(10)]
Y1=sorted(Y)
Y2=[n/2 for n in Y1]
spobj2.set_xlabel("Time in Seconds")
spobj2.set_ylabel("Current in A")
spobj2.set_title("Output Characteristics")
spobj2.plot(X,Y1,color='b',marker='*',linestyle='-',mec='g',mfc='r',label='Full load')
spobj2.plot(X,Y2,color='c',marker='v',linestyle=':',mec='r',mfc='k',label='Half Load')
spobj2.legend()
```
