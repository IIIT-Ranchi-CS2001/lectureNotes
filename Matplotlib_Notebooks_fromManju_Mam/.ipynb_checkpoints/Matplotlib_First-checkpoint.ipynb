facecolor takes hex value for giving color back ground with format #RRGGBB ( 8 bits for indicating  intensity each primary color)
#000000  indicates black, #FFFFFF indicates white, #FF0000 --> RED, #00FF00 --> Green etc 

plt.axes() helps to set X and Y axes whose limit ranges from 0 to 1 (default limits)

```python
plt.figure(figsize=(6,6),facecolor='0.5') # facecolor is grayscale value ranging from 0 to 1
# facecolor='0' implies black background and '1' indicates white background
ax=plt.axes()
ax.set_xlim(0,11)
ax.set_ylim(0,50)
```

ax is the figure object with default axes. we can set our x limit and y limit upon the X and Y axes by using the built in functions set_xlim() and set_ylim(). The input argument can be a list or a tuple also


```python
ax=plt.axes()
ax.set_xlim((0,10))
ax.set_ylim([0,50])
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
```

To customize the tickpoints of X and y axes, we can use methods set_xticks() and set_yticks()

```python
import random
ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
Y=[random.randrange(0,50,10) for n in range(10)]
ax.plot(Y)
```

To plot a simple line graph, a list of random numbers within the range 0 to 50 is created. To plot line graph, method used is plot(). plot can take  a number of arguments to add various features for the graph. Values on Y axis is mandatory. If X and Y axis values are given, both shall be of equal length.

```python

ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(5)] # X is having only 5 values
Y=[random.randrange(0,50,10) for n in range(10)]

ax.plot(X,Y)
```

```python
ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] # X takes 10 values
Y=[random.randrange(0,50,10) for n in range(10)]# Y takes 10 values
Y.sort() # sorting the random numbers to avoid zig zag shape of the graph
ax.plot(X,Y)
```

#Important features of plot

        linestyle or ls: {'-', '--', '-.', ':', '', (offset, on-off-seq), ...}
        linewidth or lw: float
        marker: marker style string, `~.path.Path` or `~.markers.MarkerStyle`
        markeredgecolor or mec: color
        markeredgewidth or mew: float
        markerfacecolor or mfc: color
        markerfacecoloralt or mfcalt: color
        markersize or ms: float
valid options for each argument

```python
ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,10) for n in range(10)]
Y.sort()
ax.plot(X,Y,color='r',marker='o',linestyle='-',mec='g',mfc='b')
```

```python
ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,10) for n in range(10)]
Y.sort()
ax.plot(X,Y,color='r',marker='o',ls='-',lw=4, mec='g',mfc='b') # ls=line style, lw= line width
```

```python
# to annotate the X and Y coordinates of the marker points. 
ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,4) for n in range(10)]
Y.sort()
for x,y in zip(X,Y):
 ax.annotate('('+str(x)+','+str(y)+')',xy=(x,y))
ax.plot(X,Y,color='r',marker='o',linestyle='-',mec='g',mfc='b')
```

```python
# to annotate points on the graph and also to add some text data onto a specific position in the graph
ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,4) for n in range(10)]
Y.sort()
for x,y in zip(X,Y):
 ax.annotate('('+str(x)+','+str(y)+')',xy=(x,y))
ax.plot(X,Y,color='r',marker='o',linestyle='-',mec='g',mfc='b')
ax.annotate("My first attempt",xy=(3,10))
```

```python
 # to plot multiple graphs on the same axes

ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,10) for n in range(10)]
Y1=sorted(Y)
Y2=[n*2 for n in Y1]
ax.plot(X,Y1,color='r',marker='o',linestyle='-',mec='g',mfc='b',label='Once')
ax.plot(X,Y2,color='g',marker='D',linestyle='-',mec='b',mfc='r',label='Twice')
ax.legend()
```

```python

ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,10) for n in range(10)]
Y1=sorted(Y)
Y2=[n*2 for n in Y1]
ax.set_xlabel("Time in Seconds")
ax.set_ylabel("Amplitude in Volts")
ax.set_title("Sample Graphs")
ax.plot(X,Y1,color='r',marker='o',linestyle='-',mec='g',mfc='b',label='Once')
ax.plot(X,Y2,color='g',marker='D',linestyle=':',mec='b',mfc='r',label='Twice')
ax.legend()
```

```python

ax=plt.axes()
ax.set_xlim(0,10)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,11)])
y1=ax.set_yticks([n for n in range(0,51,5)])
X=[n for n in range(10)] 
Y=[random.randrange(0,50,10) for n in range(10)]
Y1=sorted(Y)
Y2=[n*2 for n in Y1]
ax.set_xlabel("Time in Seconds")
ax.set_ylabel("Amplitude in Volts")
ax.set_title("Sample Graphs")
ax.plot(X,Y1,'ro-',mec='g',mfc='b',label='Once',drawstyle='steps')
ax.plot(X,Y2,'gD:',mec='b',mfc='r',label='Twice',drawstyle='steps')
ax.legend()
```

```python
import random
ax=plt.axes()
ax.set_xlim(0,5)
ax.set_ylim(0,50)
x1=ax.set_xticks([n for n in range(0,5)])
y1=ax.set_yticks([n for n in range(0,51,10)])
ax.set_xticklabels(['Phase 0', 'Phase 1', 'Phase 2', 'Phase3', 'Phase4'],rotation=90, fontsize='large')
ax.set_yticklabels(['LOW', 'MEDIUM', 'AVERAGE', 'HIGH', 'VERY HIGH','ERROR'],rotation=0, fontsize='medium')

X=[n for n in range(5)] 
Y=[random.randrange(0,50,10) for n in range(5)]
Y1=sorted(Y)
Y2=[n*2 for n in Y1]
ax.set_xlabel("Time in Seconds")
ax.set_ylabel("Amplitude in Volts")
ax.set_title("Sample Graphs")
ax.plot(X,Y1,'ro-',mec='g',mfc='b',label='Once',drawstyle='steps')
ax.plot(X,Y2,'gD:',mec='b',mfc='r',label='Twice',drawstyle='steps')
ax.legend()
```
